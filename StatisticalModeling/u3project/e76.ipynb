```python
import numpy as np
```

```python
# first leslie model
leslieArray = np.array([[0.125, 6], [0.2, 0]])
leslieArray
```

```python
# function to generate time steps of population history of age classes
def generate_history(timeSteps, ageClassStart, leslieMatrix = leslieArray):
    ageClassHistory = []
    ageClassHistory.append(ageClassStart)
    for ageStep in range(timeSteps):
        newAgeClass = np.matmul(leslieMatrix, ageClassHistory[ageStep])
        ageClassHistory.append(newAgeClass)
    return ageClassHistory
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ageClassStartTest = np.array([[10], [990]])
historyTest = generate_history(10, ageClassStartTest)
historyTest
```

```python
import pandas as pd
```

```python
def plot_history(history):
    immature = []
    adult = []
    for ageClass in history:
        immature.append(ageClass[0][0])
        adult.append(ageClass[1][0])
    
    historyData = pd.DataFrame({"Immature": immature, "Adult": adult})
    ax = sns.lineplot(historyData, markers=True, dashes=False)
    ax.set(title = "Population development")
    for x, y in zip(range(len(immature)), immature):
        plt.text(x, y, "{:.0f}".format(y))
    for x, y in zip(range(len(adult)), adult):
        plt.text(x, y, "{:.0f}".format(y))
    
```

```python
plot_history(historyTest)
```

```python
ageClassStartTest1 = np.array([[83760], [143280]])
historyTest1 = generate_history(20, ageClassStartTest1)
plt.figure(figsize = (10, 10))
plot_history(historyTest1)
```

```python
evalues, evectors = np.linalg.eig(leslieArray)
print(evalues)
print(evectors)
plot_history(generate_history(10, evectors[0].reshape(-1,1)))
```

```python
plot_history(generate_history(10, evectors[1].reshape(-1,1)))
```

```python
def e76Repeat(leslieMatrix):
    plt.figure(figsize = (20, 5))
    evalues, evectors = np.linalg.eig(leslieMatrix)
    print(evalues)
    print(evectors)
    plt.subplot(1,3,1)
    plot_history(generate_history(10, np.array([[10], [10]])))
    plt.subplot(1,3,2)
    plot_history(generate_history(10, evectors[0].reshape(-1,1)))
    plt.subplot(1,3,3)
    plot_history(generate_history(10, evectors[1].reshape(-1,1)))
```
