## Data visualization
### ZN

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python

sns.displot(x=bostonData.ZN, kde = True, rug = True, stat = 'probability')
```

```python
sns.displot(x=np.log1p(bostonData.ZN), kde = True, rug = True, stat = 'probability')
```

### ZN analysis
Most of the **ZN** values are 0
The remainder of these values seems to range from 12.50 to 100, although there seems to be some spikes around 20 and 80
Applying the logarithm makes the sections of values a lot more readable, it summarizes them into a smaller range of numbers and makes it easier. Also makes it easier to appreciate the gap between 0 and the next smallest value, 12.50
Overall, most of the **ZN** values are 0, but the remainder of the values itself skews to the left to the smallest values as well.

### INDUS

```python
sns.displot(x=bostonData.INDUS, kde = True, rug = True, stat = 'probability')
```

```python
sns.displot(x=np.log1p(bostonData.INDUS), kde = True, rug = True, stat = 'probability')
```

### INDUS analysis
The peak on the right side is because of the 18.100 value that repeats a lot, but the range from 5 to 9 is also decently common. The normalized version makes it easier to see that the skew is to the right side by quite a lot.

### Pairwise relationships

### Correlations
Negative: TARGET-LSTAT, RM-LSTAT, AGE-DIS, NOX-DIS, DIS-INDUS

Conclusions could still be drawn, but careful analysis needs to be done, as we can't be completely sure about them.
