```python
# Replaced ? with nan
dataNan = data.replace('?', np.NaN)
```

```python
dataNan
```

```python
avgNl = dataNan["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avgNl)
```

```python
# Fix values in normalized-losses
dataNan["normalized-losses"].replace(np.nan, avgNl, inplace=True)
```

```python
dataNan
```

```python
#Check which columns have Nan
dataNan.isnull().sum()
```

```python
# Find mode of number of doors to use as replacement
dataNan["num-of-doors"].mode()
```

```python
dataNan["num-of-doors"].replace(np.nan, 'four', inplace = True)
```

```python
dataNan.isnull().sum()
```

```python
avgBore = dataNan["bore"].astype("float").mean(axis=0)
print("Average of bore:", avgBore)
```

```python
avgBore = '%.2f'%(float(avgBore))
print(avgBore)
```

```python
avgBore
```

```python
dataNan["bore"].replace(np.nan, float(avgBore), inplace = True)
```

```python
dataNan.isnull().sum()
```

```python
avgStroke = '%.2f'%dataNan["stroke"].astype("float").mean(axis=0)
print("Average of stroke:", avgStroke)
```

```python
dataNan["stroke"].replace(np.nan, float(avgStroke), inplace = True)
```

```python
dataNan.isnull().sum()
```

```python
dataNan["horsepower"].fillna(dataNan["horsepower"].astype(float).mean(axis=0), inplace=True)
```

```python
dataNan.isnull().sum()
```

```python
dataNan["peak-rpm"].fillna(dataNan["peak-rpm"].astype(float).mean(axis=0), inplace=True)
```

```python
dataNan["price"].fillna(dataNan["price"].astype(float).mean(axis=0), inplace=True)
```

```python
dataNan.isnull().sum()
```
